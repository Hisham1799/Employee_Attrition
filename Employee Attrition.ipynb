```python
# Let's visualize how attrition is related with each and every variable

#importing Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# List of categorical columns
categorical_cols = [
    "JobRole", "MaritalStatus", "OverTime", "BusinessTravel", 
    "Department", "EducationField", "Gender", "JobLevel"
]

# Loop through categorical columns and plot attrition rate by each category
for col in categorical_cols:
    # Calculate the attrition rate for each category in the column
    attrition_rate_data = (
        df.groupby(col)["Attrition"]
        .value_counts(normalize=False)  # Counts of 'Yes' and 'No'
        .unstack(fill_value=0)         # Pivot the table to have 'Yes' and 'No' as columns
        .apply(lambda x: (x['Yes'] / (x['Yes'] + x['No'])) * 100, axis=1)  # Calculate attrition rate
        .reset_index(name="Attrition Rate (%)")  # Reset index to create a DataFrame
    )

    # Plot attrition rate by category
    plt.figure(figsize=(12, 6))
    sns.barplot(data=attrition_rate_data, x=col, y="Attrition Rate (%)", palette="Set2")
    plt.title(f"Attrition Rate by {col}")
    plt.ylabel("Attrition Rate (%)")
    plt.xticks(rotation=45)
    plt.show()
```

# Building Machine Learning Model

   - LogisticRegression

---

## Overview of Machine Learning Models

We are using three different methods to help the computer make predictions based on data. Here's a brief overview:

### 1. **Random Forest Classifier**
   - This method uses multiple decision trees. Each tree gives an answer, and the most common answer from all trees is chosen.
   - It is great at handling complex data and is often very accurate.

### 2. **Logistic Regression**
   - This method predicts the chance of something happening. For example, it might predict the likelihood of a employee leaving the company.
   - It is simple and fast but may not be as good with very complex data.

### 3. **K-Nearest Neighbors (KNN)**
   - This method looks at the closest examples to the data point it’s trying to predict, and makes a decision based on the majority of those nearby points.
   - It’s simple but can be slow when there’s a lot of data.

### 4. **Comparing Their Performance**
   - After building these models, we test them to see which one makes the best predictions.
   - The model with the most accurate predictions is chosen for use.

---


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt

# Encode the 'Attrition' column
le = LabelEncoder()
df['Attrition_Encoded'] = le.fit_transform(df['Attrition'])

# Separate the classes
df_no = df[df['Attrition_Encoded'] == 0]
df_yes = df[df['Attrition_Encoded'] == 1]

# Eliminate of 'Attrition = No' rows (keep only 37%)
df_no_sampled = df_no.sample(frac=0.37, random_state=36)

# Combine the sampled 'Attrition = No' with all 'Attrition = Yes'
df_balanced = pd.concat([df_yes, df_no_sampled])

# Shuffle the combined DataFrame
df_balanced = df_balanced.sample(frac=0.3, random_state=93).reset_index(drop=True)

# Select features and target
features = df_balanced.drop(['Attrition', 'Attrition_Encoded'], axis=1)
target = df_balanced['Attrition_Encoded']

# Encode categorical columns
features = pd.get_dummies(features, drop_first=True)

# Scale the numerical features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Split the data into training and testing sets using Stratified Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42, stratify=target)

# Apply SMOTE to balance the classes in the training set
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Initialize models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    'Logistic Regression': LogisticRegression(random_state=42, class_weight='balanced'),
    'KNN': KNeighborsClassifier(n_neighbors=5),
}

# Evaluate models
for model_name, model in models.items():
    print(f"Evaluating {model_name}...")
    model.fit(X_train_res, y_train_res)  # Train the model
    y_pred = model.predict(X_test)  # Predict on the test set
    
    # Print classification report
    print(classification_report(y_test, y_pred))
    
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```


---

## Model Evaluation Summary

We tested three different models to make predictions based on data. Here’s a simple breakdown of how each performed:

### Key Metrics Explained:
- **Accuracy**: How often the model gets things right. A higher number means the model is more accurate overall.
- **Precision**: How often the model is correct when it predicts something (for example, predicting whether an employee will leave the company). Higher precision means fewer mistakes.
- **Recall**: How good the model is at finding all the correct answers (for example, identifying all the employees who are likely to leave the company). Higher recall means fewer important things are missed.
- **F1-Score**: This combines precision and recall into one number. The higher the F1-Score, the better the model is at both being accurate and not missing anything important.

---

### 1. **Random Forest Model**
   - **Accuracy**: 69% – The model got 69% of predictions correct.
   - **Precision**: 76% for one group, and 59% for another.
     - This means it was more accurate in predicting one group correctly, but made more mistakes in the other group.
   - **Recall**: 73% for one group, and 62% for the other group.
     - The model was better at finding the first group but not as good at finding the second group.
   - **F1-Score**: 75% for the first group and 61% for the second group.
     - This means the model performed well in balancing accuracy and finding the right things for the first group, but less so for the second group.

   **Conclusion**: The Random Forest model did well overall, especially for one group, but could be improved for the other.

---

### 2. **Logistic Regression Model**
   - **Accuracy**: 69% – The same as Random Forest.
   - **Precision**: 74% for one group, and 60% for the other.
     - Similar to Random Forest, but it performed a bit better with the first group.
   - **Recall**: 77% for one group, and 56% for the other.
     - It was better at finding the first group, but it missed some of the second group.
   - **F1-Score**: 75% for the first group, and 58% for the second group.
     - A good balance for the first group, but less effective for the second.

   **Conclusion**: This model performed similarly to Random Forest but wasn’t as good at finding all of the second group.

---

### 3. **K-Nearest Neighbors (KNN) Model**
   - **Accuracy**: 52% – The lowest of all models.
   - **Precision**: 65% for one group, and 41% for the other.
     - The model was better with the first group but made a lot of mistakes with the second.
   - **Recall**: 50% for one group, and 56% for the other.
     - It wasn’t great at finding the right answers, especially for the first group.
   - **F1-Score**: 57% for the first group, and 47% for the second.
     - Overall, this model struggled to balance accuracy and recall.

   **Conclusion**: The KNN model performed the worst, making it less reliable for this task.

---

### Final Recommendation:
- **Random Forest** is the best model overall, as it provides a good balance of accuracy and performance, especially for the first group.
- **Logistic Regression** also performed well but was not as effective at identifying all the right answers for the second group.
- **K-Nearest Neighbors (KNN)** should be avoided because it didn’t perform as well as the other two models.

We recommend using **Random Forest** for the most accurate and reliable predictions on employee attrition.
